```python
import pandas as pd

series_apples = pd.Series([1, 3, 7, 4])
series_bananas = pd.Series([2, 6, 3, 5])

df = pd.DataFrame({ 'Apples': series_apples, 'Bananas': series_bananas })

print(df)
```

```python
a = ["Google", "Runoob", "Taobao"]
myVar = pd.Series(a, index=["x", "y", "z"])
display(myVar)
```

```python
data = [1, 2, 3, 4, 5]
index = ['a', 'b', 'c', 'd', 'e']
myVar = pd.Series(data, index=index)

display(myVar.index)
display(myVar.values)
display(myVar.dtype)
print(myVar.head(2))

cumsum = myVar.cumsum()
display(cumsum)
```
